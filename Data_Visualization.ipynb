# Subplot

```python
df.columns
```

```python
num_cols=df.select_dtypes(include="number")
```

```python
plt.figure(figsize=(20,5))
for index,col in enumerate(num_cols.columns,start=1):
  plt.subplot(2,4,index)
  sns.boxplot(x=num_cols[col])
plt.tight_layout(w_pad=5,h_pad=3)
plt.show()
```

```python
plt.figure(figsize=(20,5))
for index,col in enumerate(num_cols.columns,start=1):
  plt.subplot(2,4,index)
  sns.histplot(x=num_cols[col])
plt.tight_layout(w_pad=5,h_pad=3)
plt.show()
```

```python
plt.figure(figsize=(20,5))
for index,col in enumerate(num_cols.columns,start=1):
  plt.subplot(2,4,index)
  sns.scatterplot(x=num_cols[col],y=num_cols["total_bill"])
plt.tight_layout(w_pad=5,h_pad=3)
plt.show()
```
